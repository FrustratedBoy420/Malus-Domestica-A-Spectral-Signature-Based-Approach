```python
from glob import *
import spectral
import matplotlib.pyplot as plt
import os
```

```python
def create_folder(folder_path):
  if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    print(f"Folder '{folder_path}' created.")
  else:
    print(f"Folder '{folder_path}' already exists.")

def files(filelist):
  hdr_list=[]
  img_list=[]
  for i in filelist:
    if i.endswith('.img'):
      img_list.append(i)
    else:
      hdr_list.append(i)
  return hdr_list,img_list
def show_img(hdr_list,img_list,outputfolder):
  for i in range(len(hdr_list)):
    img=spectral.envi.open(hdr_list[i],img_list[i])
    img_data=img.load()
    view=spectral.imshow(img_data)
    save_path = os.path.join(outputfolder, f"{hdr_list[i].split('/')[-1].split('.')[0]}.JPG")
    plt.savefig(save_path)
    plt.show()


```

```python
healthy_folder_img = 'Dataset/Valid/Healthy'
infected_folder_img = 'Dataset/Valid/Infected'
```

```python
create_folder(healthy_folder_img)
create_folder(infected_folder_img)
```

```python
healthy_folder=glob("Data Files/serie healthy leaves/*")
healthy_folder.sort()
infected_folder=glob("Data Files/serie infected leaves/*")
infected_folder.sort()
```

```python
healthy_hdr_list,healthy_img_list=files(healthy_folder)
infected_hdr_list,infected_img_list=files(infected_folder)
```

```python
show_img(healthy_hdr_list,healthy_img_list,healthy_folder_img)
show_img(infected_hdr_list,infected_img_list,infected_folder_img)
```
